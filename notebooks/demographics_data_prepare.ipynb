```python
df = pd.read_excel("../data/demographics_data/sp_1.2.xlsx", sheet_name="Субъекты")
df = df.iloc[3:]

df = df.rename(columns={'Unnamed: 0': 'region'})
new_columns = ['region']


for i in range(1, len(df.columns), 2):
    if i + 1 >= len(df.columns):
        break
    year = df.columns[i].split()[0]
    new_columns.extend([f'men_{year}', f'women_{year}'])

df.columns = new_columns

years = sorted(set(int(col.split('_')[1]) for col in df.columns if 'men_' in col), reverse=True)
sorted_columns = ['region']
for year in years:
    sorted_columns.extend([f'men_{year}', f'women_{year}'])

df = df[sorted_columns]

df_long = pd.wide_to_long(
    df,
    stubnames=['men', 'women'],
    i='region',
    j='year',
    sep='_',
    suffix='\d+'
).reset_index()

df_long = df_long.sort_values('year')
df_long = df_long[['year', 'region', 'men', 'women']]
df_long.to_csv("sp12.csv", index=False)
```

```python
df = pd.read_excel("../data/demographics_data/sp_1.1.xlsx", sheet_name="Числ. нас. субъекты")
df = df.iloc[3:]

df = df.rename(columns={'Unnamed: 0': 'region'})
new_columns = ['region']
import re
import pandas as pd

# Получаем текущие названия колонок
current_columns = df.columns.tolist()

# Инициализируем новые названия с 'region'
new_columns = ['region']

# Обрабатываем колонки, начиная с индекса 1, группируя по три
for i in range(1, len(current_columns), 3):
    group = current_columns[i:i+3]
    # Извлекаем год из первого элемента группы
    year_str = str(group[0])
    year = re.search(r'\d{4}', year_str).group()
    # Генерируем новые названия
    new_names = [
        f'population_{year}',
        f'employable_population_{year}',
        f'employable_population_share_{year}'
    ]
    new_columns.extend(new_names)

# Убедимся, что количество колонок совпадает
if len(new_columns) != len(current_columns):
    raise ValueError("Несоответствие количества колонок после преобразования")

# Применяем новые имена
df.columns = new_columns

# Преобразуем в длинный формат
df_long = pd.wide_to_long(
    df,
    stubnames=['population', 'employable_population', 'employable_population_share'],
    i='region',
    j='year',
    sep='_',
    suffix='\d+'
).reset_index()

# Упорядочим колонки
df_long = df_long[['year', 'region', 'population', 'employable_population', 'employable_population_share']]

# Сортируем по региону и году (опционально)
df_long = df_long.sort_values(['region', 'year'])
df_long.to_csv("sp11.csv", index=False)
```

```python
# Дополним словарь для специфических случаев из ваших данных
additional_mapping = {
    'г.Москва': 'Moscow and Moscow region',
    'г.Санкт-Петербург': 'St. Petersburg',
    'Oмская область': 'Omsk region',  # Исправление латинской 'O'
    'Кемеровская область – Кузбасс': 'Kemerovo region',
    'Республика Северная Осетия – Алания': 'Republic of North Ossetia - Alania',
    '   Ненецкий автономный округ': 'Nenets Autonomous Okrug',
    '   Ханты-Мансийский автономный округ – Югра': 'Khanty-Mansi Autonomous Okrug',
    '   Ямало-Ненецкий автономный округ': 'Yamalo-Nenets Autonomous Okrug'
}

# Объединяем основной словарь с дополнениями
full_rename_dict = {**rename_dict, **additional_mapping}

# Функция для очистки названий
def clean_region_name(name: str) -> str:
    return (
        name.strip()
        .replace('–', '-')  # Заменяем длинное тире
        .replace('  ', ' ')  # Убираем двойные пробелы
        .replace('авт. округа', 'автономный округ')
    )

df_long_1 = pd.read_csv("sp12.csv")

# Применяем преобразования к колонке region
df_long_1['region'] = (
    df_long_1['region']
    .apply(clean_region_name)
    .replace(full_rename_dict)
)

# Проверяем результат
print("Уникальные регионы после преобразования:")
print(df_long_1['region'].unique())
```

```python
from linearmodels.panel import PanelOLS, RandomEffects
from linearmodels.panel.results import compare
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.diagnostic import het_breuschpagan

# 1. Подготовка данных
df = df_final.set_index(['region', 'year']).copy()  # Установка индексов панельных данных
df['constant'] = 1  # Константа для LSDV

# 2. LSDV модель (фиктивные переменные для регионов)
formula = 'geometric_mean ~ employable_population_share + men + women + C(region) - 1'
lsdv_model = smf.ols(formula, data=df.reset_index()).fit()
print("LSDV Model Results:")
print(lsdv_model.summary())

fe_model = PanelOLS(
    df['geometric_mean'],
    df[['employable_population_share', "men", "women"]],
    entity_effects=True,
    time_effects=True  # Если нужно учесть временные эффекты, установите True
).fit(cov_type='clustered', cluster_entity=True)
print("\nFixed Effects Model Results:")
print(fe_model)

re_model = RandomEffects(
    df['geometric_mean'],
    df[['employable_population_share', "men", "women"]]
).fit()
print("\nRandom Effects Model Results:")
print(re_model)

comparison = compare({'FE': fe_model, 'RE': re_model})
print("\nModel Comparison with Hausman Test:")
print(comparison)
```
