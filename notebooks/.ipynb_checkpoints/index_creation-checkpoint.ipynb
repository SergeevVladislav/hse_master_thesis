```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

plt.style.use('ggplot')

df_final = pd.DataFrame()
```

# T1 Institutional data

```python
t1 = pd.read_csv("../data/institutional_data/nps_institutional_data.csv")
t1['date'] = pd.to_datetime(t1['date'], format="%d.%m.%Y")
t1.region = t1.region.replace(
    {"Республика Крым**": "Республика Крым",
     "г. Севастополь**": "г. Севастополь",
     "Кемеровская область-Кузбасс": "Кемеровская область",
     "Кемеровская область - Кузбасс": "Кемеровская область",
     "Чукотский авт. округ": "Чукотский автономный округ",
     "Еврейская авт. область": "Еврейская автономная область",
     "Республика Северная Осетия - Алания": "Республика Северная Осетия-Алания"
    }
)
t1["Total number of banking system institutions"] = t1["Total number of banking system institutions"].str.replace(" ", "").astype(float)
t1["additional offices"] = t1["additional offices"].str.replace(" ", "").astype(float)
t1["Number of bank institutions per 1 million inhabitants"] = t1["Number of bank institutions per 1 million inhabitants"].str.replace("-", "0")
t1["Number of bank institutions per 1 million inhabitants"] = t1["Number of bank institutions per 1 million inhabitants"].str.replace(" ", "").astype(float)
t1["Number of bank institutions per 1 million inhabitants"] = t1["Number of bank institutions per 1 million inhabitants"].replace({0: t1["Number of bank institutions per 1 million inhabitants"].mean()})
t1.head()
```

# T2 Accounts data

```python
t2 = pd.read_csv("../data/accounts_data/accounts_data.csv")
t2['date'] = pd.to_datetime(t2['date'], format="%d.%m.%Y")
t2.region = t2.region.replace(
    {"Республика Крым**": "Республика Крым",
     "г. Севастополь**": "г. Севастополь",
     "Кемеровская область-Кузбасс": "Кемеровская область",
     "Кемеровская область - Кузбасс": "Кемеровская область",
     "Еврейская авт. область": "Еврейская автономная область",
     "Республика Северная Осетия - Алания": "Республика Северная Осетия-Алания"
    }
)
t2["total_accounts"] = t2["total_accounts"].str.replace(" ", "").str.replace(",", ".").astype(float)
t2["total_in_credit_org"] = t2["total_in_credit_org"].str.replace(" ", "").str.replace(",", ".").astype(float)
t2["clients-legal entities that are not credit institutions"] = t2["clients-legal entities that are not credit institutions"].str.replace(" ", "").str.replace(",", ".").astype(float)
t2["individual clients"] = t2["individual clients"].str.replace(" ", "").str.replace(",", ".").astype(float)
t2["The number of accounts opened by institutions of the banking system, per 1 resident, units."] = t2["The number of accounts opened by institutions of the banking system, per 1 resident, units."].str.replace(" ", "").str.replace(",", ".").str.replace("-", "0").astype(float)
t2["The number of accounts opened by institutions of the banking system, per 1 resident, units."] = t2["The number of accounts opened by institutions of the banking system, per 1 resident, units."].replace({0: t2["The number of accounts opened by institutions of the banking system, per 1 resident, units."].mean()})
t2.head()
```

```python
# for col in df.drop(labels=["region", "date"], axis=1).columns:
#     df[f'{col}_normalized'] = (
#         df.groupby(['date'])[col]
#         .transform(lambda x: (x - x.min()) / (x.max() - x.min()))
#     )

import numpy as np

for col in df.drop(labels=["region", "date"], axis=1).columns:
    df[f'{col}_normalized'] = (
        df.groupby(['date'])[col]
        .transform(
            lambda x: (lambda tx: (tx - tx.min()) / (tx.max() - tx.min() + 1e-8))(  # +1e-8 чтобы избежать деления на 0
                np.where(x >= 1, np.log(x), 0)  # Если x < 1 → 0, иначе log(x)
            )
        )
    )
```

```python
def geometric_mean_no_zeros(row):
    # Отбираем только значения больше нуля
    positive_values = row[row > 0]
    if len(positive_values) == 0:  # Если все значения равны нулю
        return 0
    return gmean(positive_values)

df['geometric_mean'] = df[df.columns[df.columns.str.contains("_normalized")]].apply(geometric_mean_no_zeros, axis=1)
```

```python
import matplotlib.pyplot as plt

df = df.rename(
    columns={
        "bank orders_normalized": "volume bank orders",
    "The number of accounts opened by institutions of the banking system, per 1 resident, units._normalized": "accounts per 1",
    "Number of bank institutions per 1 million inhabitants_normalized": "bank_isnt_per_1",
    "Total number of banking system institutions_normalized": "total_bank",
    "volume orders in electronic form with internet_normalized": "vol_internet",
    "payment requirements with internet_normalized": "pay_internet"
    }
)

factors = ["volume bank orders", "accounts per 1", "bank_isnt_per_1", "total_bank", "vol_internet", "pay_internet"]
df_min = df[(df["date"].dt.year == 2023) & (df["date"].dt.month == 10)].sort_values(by="geometric_mean", ascending=False)
df_mosc = df[df["region"] == "Камчатский край"][factors]
df_st = df[df["region"] == "Саратовская область"][factors]

N = len(factors)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1]  # Замыкаем круг



values =df_mosc.mean().values.tolist()
values += values[:1]

values_st =df_st.mean().values.tolist()
values_st += values_st[:1]

# Инициализируем график
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'polar': True})

# Рисуем линию
ax.plot(angles, values, color='blue', linewidth=2, label='Kamchatka Krai')
ax.plot(angles, values_st, color='green', linewidth=2, label='Saratov Oblast')



# Настраиваем оси
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_rlabel_position(0)

# Добавляем подписи
plt.legend(loc='best')
plt.xticks(angles[:-1], factors)
plt.title('Middle-class Regions', y=1.08)
plt.ylim(0, 1 * 1.1)
```
