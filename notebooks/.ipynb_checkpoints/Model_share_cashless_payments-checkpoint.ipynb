```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(data, target_column=None):
    # Исключаем целевую переменную (если есть)
    if target_column:
        features = data.drop(labels=target_column, axis=1)
    else:
        features = data.copy()
    
    # Добавляем константу для расчета VIF
    features_with_const = sm.add_constant(features)
    
    # Считаем VIF для каждой переменной
    vif = pd.DataFrame()
    vif["Variable"] = features_with_const.columns
    vif["VIF"] = [variance_inflation_factor(features_with_const.values, i) 
                  for i in range(features_with_const.shape[1])]
    
    # Удаляем строку с константой (ее VIF не интерпретируется)
    vif = vif[vif["Variable"] != "const"]
    return vif


def remove_high_vif(data, threshold=5, target_column=None):
    if target_column:
        features = data.drop(labels=[target_column], axis=1)
    else:
        features = data.copy()
    
    while True:
        vif_df = calculate_vif(features, target_column)
        max_vif = vif_df["VIF"].max()
        if max_vif <= threshold:
            break
        # Находим переменную с максимальным VIF
        max_vif_row = vif_df[vif_df["VIF"] == max_vif]
        feature_to_remove = max_vif_row["Variable"].values[0]
        # Удаляем ее
        features = features.drop(columns=[feature_to_remove])
    
    return features

# Предположим, у вас есть DataFrame `df` с переменными и (опционально) целевой колонкой 'target'
cleaned_data = remove_high_vif(df.drop(labels=["value"] + cols_to_drop, axis=1), threshold=10, target_column=None)

# Вывод оставшихся переменных
print("Оставшиеся переменные:", cleaned_data.columns.tolist())
```
