```python
share = pd.DataFrame()

genders = ["B"]
groups = ["T"]
for ss in genders:
    for gg in groups:
        df_1 = df[(df["Sex"] == ss) & (df["Group"] == gg)].drop(labels=["Group", "Sex"], axis=1).rename(columns={col: col + f"_{ss}_{gg}" for col in df.columns if "PopDa" in col})
        df_f = df_1.set_index(["Year", "Reg"]).div(df_1.set_index(["Year", "Reg"]).sum(1), 0).reset_index()
        if share.empty:
            share = df_f
        else:
            share = pd.merge(df_f, share, on=["Year", "Reg"], how="inner")

        print(share.shape)
```

```python
a = {}
for key in region_code_to_name.keys():
    a[str(key)] = region_code_to_name[key]
```

```python
df["Reg"] = df["Reg"].astype(str).map(a)
df = df.dropna()
```

```python
df = df[df["Sex"] != "B"]
df = df[df["Group"] != "T"]
```

```python
df = df.drop(labels=["PopD5a0", "PopD5a1", "PopD5a5", "PopD5a10", "PopD5a15"], axis=1)
df.columns
```

```python
import pandas as pd

pop_cols = [c for c in df.columns if c.startswith("PopD5a")]

df_wide = df.pivot_table(
    index=["Year","Reg"],
    columns=["Group","Sex"],
    values=pop_cols,
    aggfunc="sum"
)

df_wide.columns = [
    f"{pop}_{grp}_{sex}"
    for pop, grp, sex in df_wide.columns
]

df_wide = df_wide.reset_index()

df_wide.head()
```

```python
df_wide.to_csv("../data/hse_data/nes.csv", index=False)
```

```python
import pandas as pd

# 1) Список всех PopD5a-колонок
pop_cols = [c for c in df.columns if c.startswith("PopD5a")]

# 2) Строим сводную таблицу с абсолютными значениями
pivot_abs = df.pivot_table(
    index=["Year", "Reg"],
    columns=["Group", "Sex"],
    values=pop_cols,
    aggfunc="sum"    # суммирует дубликаты
)

# 3) «Сплющиваем» многомерные имена колонок в формат PopD5aX_<Group>_<Sex>
pivot_abs.columns = [
    f"{pop}_{grp}_{sex}"
    for pop, grp, sex in pivot_abs.columns
]

# 4) Для каждого (Year, Reg) считаем общий итого-по-строке
row_totals = pivot_abs.sum(axis=1)

# 5) Делим каждую ячейку на соответствующий итого и получаем доли
pivot_frac = pivot_abs.div(row_totals, axis=0)

# 6) Возвращаем Year и Reg в столбцы
pivot_frac = pivot_frac.reset_index()

print(pivot_frac.head())
```
